```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_name = 'Airplane_Crashes_and_Fatalities_Since_1908.csv'
file_path_full = os.path.join(path, file_name)


df = pd.read_csv(file_path_full)

df['Time'].fillna('Unknown', inplace=True)
df['Location'].fillna('Unknown', inplace=True)
df['Operator'].fillna('Unknown', inplace=True)
df['Flight #'].fillna('Not Applicable', inplace=True)
df['Route'].fillna('Unknown', inplace=True)
df['Type'].fillna('Unknown', inplace=True)
df['Registration'].fillna('Not Applicable', inplace=True)
df['cn/In'].fillna('Not Applicable', inplace=True)
df['Aboard'].fillna(df['Aboard'].median(), inplace=True)
df['Fatalities'].fillna(df['Fatalities'].median(), inplace=True)
df['Ground'].fillna(df['Ground'].median(), inplace=True)
df['Summary'].fillna('No summary available', inplace=True)

print("Missing data after imputation:")
print(df.isnull().sum())
```
